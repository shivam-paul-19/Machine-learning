# **Cross Validation**

Cross-validation is a robust resampling technique used in machine learning to assess how well a model generalizes to unseen data, preventing overfitting by training and testing on different subsets of the data iteratively. The process splits data into $k$ parts (folds), training on $k-1$ and testing on the remaining part repeatedly, then averaging the results.Â 

## Types of Cross validation

- ### **Leave one out CV**
    In this cross validation technique, at every experiment, only **one** instance of training data is taken as *validation data*.\ 
    For eg, if the training data has $1000$ data points then in the first experiment, the $0_{th}$ data point is taken as validation data and the model will be trained on the rest $999$ data points. In second experiment, $1_{st}$ data point is taken as validation data... this process goes on.\
    This method has a limitation that it is very time consuming, because for 1000 data points, every single time one data instance is taken. Also the size of validation data is small which leads to **overfitting**.

    > Note: There's a variation of this, **Leave p out cv**, where the validation data is $p$, not one, $p$ is set by the programmer.

- ### **K-fold CV**
    In this cross validation, we set a value *k*, which determines the size of validation data and the number of experiments.
    
    $$\text{Validation data} = \frac{Train\ data}{k}$$

    For eg, if there are 300 data points, and the value of k is 5, then validation data will be 60 and there will be 5 experiments. In first experiment, first 60 records will be validation data and model will bre trained on the rest, in second experiment, next 60 will be validation data, and this goes on for *k* times.

    > It also has a variation, which is **Strstified K fold CV**, this is basically for classfication problem where it makes sure that the validation data is balanced.
