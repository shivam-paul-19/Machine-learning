The Random Forest is a powerful supervised machine learning algorithm that builds multiple decision trees during training and merges their predictions to produce a more accurate and stable result. It is widely used for both classification and regression tasks. It is basically a **Bagging** technique.\
For classification, it goes for the majority outcome out of all *decision trees*. And in Regression, it goes for the *Mean* of the outputs.

## Implementation

```python
# Get the data set

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
print(data['DESCR'])
```

```python
import pandas as pd

X, y = pd.DataFrame(data['data'], columns=data['feature_names']), data['target']
X.head()
```

```python
# split into test train data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# train the model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50)    # using 50 decision trees
rf.fit(X_train, y_train)
```

```python
# make predictions
y_pred = rf.predict(X_test)
y_pred
```

```python
# check for accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_pred=y_pred, y_true=y_test))
print(confusion_matrix(y_pred=y_pred, y_true=y_test))
print(classification_report(y_pred=y_pred, y_true=y_test))
```

## Let's try with one decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_pred=y_pred, y_true=y_test))
print(confusion_matrix(y_pred=y_pred, y_true=y_test))
print(classification_report(y_pred=y_pred, y_true=y_test))
```
