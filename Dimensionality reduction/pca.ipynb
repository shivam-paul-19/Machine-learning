Principal Component Analysis (PCA) is a popular unsupervised linear transformation technique used in machine learning to reduce the dimensionality of large datasets while retaining most of the original variance (information). It transforms high-dimensional, correlated features into a smaller set of uncorrelated "principal components". PCA is essential for speeding up training, improving visualization, and mitigating the "curse of dimensionality" and overfitting.

## Working of PCA
The Aim of PCA is to find an *Eigon Vector*, which refers to a vector that captures the highest possible spread of the data points.
Suppose the original points are given as $ P_{1}, P_{2}, ... P_{n} $ in a 2D space then the 1D representation of them will the *Projection on Eigon vector*. Let us represent the Eigon vector as $e$.

$$ P_{x}^{e} = \frac{P_{x}â‹…e}{|e|} $$
$$ \frac{e}{|e|} \text{ is basically the unit vector of e} $$
$$ P_{x}^{e} = P_{x}â‹…e_{u} $$
$$ (e_{u} = \text{unit vector of e}) $$

The values $ P_{x}^{e} $ are the new points represented in 1D, and the Eigon vector is the new X axis, which is also known as *PC1*.

Now, it is not possible to try every vector and check the variance of the projections. So we have a method of Eigon decompostion on covariance matrix.

- #### Step 1: **Normalise the Data**
    The first step is to *center the data* or normalise it.

- #### Step 2: **Find the covariance matrix**

    $$ \text{Covariance Matrix} = \begin{bmatrix} Var(x) & Cov(x, y) \\\\ Cov(y, x) & Var(y) \end{bmatrix}$$

- #### Step 3: **Find the Eigon Values**
    The Covariance matrix is represented as $A$, and the Eigon vector as $v$.

    $$ A*v = Î»*v $$
    $$ Î» = \text{Eigon values} $$

- #### Step 4: **Find the Eigon Vector**
    The vector with highest eigon value has the most spread data so that is considered as Eigon vector, and the points are transformed as the projection.

## Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

cancer_df = load_breast_cancer()
cancer_df = pd.DataFrame(cancer_df['data'], columns=cancer_df['feature_names'])
cancer_df
```

```python
# standardise the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_data = scaler.fit_transform(cancer_df)
scaled_data
```

```python
# apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(scaled_data)
pca_data
```

> All the features are reduced to 2 features now, in such a way that maximum variance is rathered.
